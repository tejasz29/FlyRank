## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

## Research Question

Which content pages should be reviewed first to maximize the impact of limited editorial effort?

### Decision

Determine which pages deserve the highest priority for content review.

### Unit of Analysis

One content page.

### Output

A ranked list of pages with a priority score, confidence level, and reason codes explaining why each page appears in the ranking.

### Action

A content manager or SEO specialist reviews the highest-ranked pages first and decides whether to refresh, expand, rewrite, monitor, or leave them unchanged.

### Cost of a Wrong Recommendation

If a page is ranked too highly, editors may spend time reviewing content that does not need attention. If an important page is ranked too low, the team may miss an opportunity to recover traffic or improve performance. Because review time is limited, ranking quality directly affects productivity.

### Why ML Helps

Many signals contribute to content performance simultaneously, including impressions, clicks, CTR, freshness, engagement, and content age. Machine learning can combine these signals into a better ranking than simple fixed rules while still supporting human decision-making.



## Quick Look at the Data

The starter dataset contains **30,000** content pages.

Out of these, **16,262 pages (54.2%)** are currently labelled as declining (`trend_direction = "down"`), indicating that there are enough positive examples to study refresh prioritization.

The median page received **731 impressions over the last 90 days**, showing that the dataset contains a meaningful amount of search activity while also including both low- and high-visibility pages.

These numbers suggest that manually reviewing every page would be inefficient, making a ranked review queue a suitable decision-support tool.

```python
import pandas as pd

df = pd.read_csv("content_refresh_anonymized.csv")  # Change the path if needed

print("Total pages:", len(df))

declining = (df["trend_direction"] == "down").sum()
print("Declining pages:", declining)
print("Declining percentage:", round(declining / len(df) * 100, 2), "%")

print("Median impressions (90d):", df["impressions_90d"].median())
print("Mean impressions (90d):", round(df["impressions_90d"].mean(), 2))
```

## 4. Careful words: what I can and can't claim


This project is designed to support decision-making rather than automate it. The ranking produced by this analysis helps identify content pages that appear to deserve review based on observable search and engagement signals.

The recommendations are based on patterns in historical data and should be treated as evidence for prioritization, not proof that a specific action will improve performance.

This project cannot prove that refreshing a page will increase traffic or engagement. It also cannot identify Google's ranking factors or establish a causal relationship between content updates and future performance. Any recommended action should be reviewed by a human before implementation.

The model uses only observable signals from the dataset and avoids relying on product decision scores or future information that could introduce data leakage. The results should therefore be interpreted as decision-support recommendations rather than guaranteed outcomes.
