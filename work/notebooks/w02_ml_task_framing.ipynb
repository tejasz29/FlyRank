## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

## Chosen Lane
Refresh / Content Opportunity Scoring

## ML Task Type
Ranking with a supervised classification model.

The goal of this project is to prioritize content pages that should be reviewed first. While the model predicts the probability that a page belongs to the "declining" class, the final output is a ranked list rather than a simple yes/no prediction.

Each page receives a score that helps content managers decide which pages deserve attention first. Therefore, classification provides the prediction while ranking is the final business objective.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*


The starter dataset uses the following proxy target:

**trend_direction == "down"**

This creates a binary target called **is_declining_label**.

The target is a proxy rather than a perfect definition of content needing refresh. In later weeks, this could be improved by defining a future-looking target such as:

Previous 90 days of features → decline during the next 30 days

Using a future outcome would better represent real-world decision making while avoiding data leakage.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The main objective is to rank pages that deserve review.

The most suitable evaluation metrics are:

- Precision@K
- Average Precision
- ROC-AUC (secondary)

Precision@K is especially important because content teams only have time to review a limited number of pages. A higher Precision@K means more of the highest-ranked recommendations are actually useful, improving the efficiency of manual review.
